# Exploratory Data Analysis (EDA)
In this notebook, we explore the Telco Customer Churn dataset to understand the distribution of features and identify key patterns related to churn.

```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Configure visual style
sns.set_theme(style='whitegrid')
```

## 1. Data Loading

```python
df = pd.read_csv('../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv')
display(df.head())
print(f'Dataset shape: {df.shape}')
```

## 2. Basic Cleaning
Handling the empty strings in `TotalCharges`.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'].replace(' ', ''), errors='coerce')
df['TotalCharges'].fillna(0, inplace=True)
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
```

## 3. Target Variable Distribution (Churn)

```python
fig = px.pie(df, names='Churn', title='Churn Distribution', color_discrete_sequence=['#2ecc71', '#e74c3c'])
fig.show()
```

We can observe a class imbalance (~26% Churn) which we will address during modeling using SMOTE or class weights.

## 4. Bivariate Analysis
Analyzing numerical features vs Churn.

```python
fig = px.box(df, x='Churn', y='tenure', title='Tenure vs Churn', color='Churn')
fig.show()
```

As expected, customers with lower tenure (newer customers) are much more likely to churn.

```python
fig = px.histogram(df, x='MonthlyCharges', color='Churn', title='Monthly Charges Distribution by Churn', barmode='overlay')
fig.show()
```

Customers with higher monthly charges show a significantly higher churn rate compared to those on basic plans.

## 5. Contract Types
Let's analyze how the type of contract affects retention.

```python
fig = px.histogram(df, x='Contract', color='Churn', barmode='group', title='Churn by Contract Type')
fig.show()
```

Month-to-month contracts are highly volatile and represent the vast majority of churned customers.
